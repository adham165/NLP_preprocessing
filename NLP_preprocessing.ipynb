```python
text_data = [
    "The movie was fantastic and I loved every part of it about Egypt",
    "I hated the film, it was the worst I have ever seen",
    "The storyline was boring but the acting was brilliant",
    "An amazing movie with a great plot and incredible performances",
    "Egypt movie, I regret wasting my time on it",
    "The actors did a great job but the story lacked depth",
    "One of the best films I have seen in a long time, highly recommend it",
    "This film was just okay, not too bad but not great either",
    "Absolutely loved the movie, fantastic plot and wonderful cast",
    "The movie was disappointing, it did not live up to the hype"
]
```

```python
tokenized_data=[]
for text in text_data:
  tokenized_data.append(nltk.tokenize.word_tokenize(text))
```

```python
stop_words = set(nltk.corpus.stopwords.words('english'))
tokenized_data_without_stopwords = []
for tokens in tokenized_data:
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    tokenized_data_without_stopwords.append(filtered_tokens)
tokenized_data_without_stopwords
```

```python
stemmed_data = []
for tokens in tokenized_data_without_stopwords:
    stemmed_tokens = [nltk.stem.PorterStemmer().stem(word) for word in tokens]
    stemmed_data.append(stemmed_tokens)
stemmed_data
```

```python
lemmetized_data = []
for tokens in tokenized_data_without_stopwords:
    lemmetized_tokens = [nltk.stem.WordNetLemmatizer().lemmatize(word) for word in tokens]
    lemmetized_data.append(lemmetized_tokens)
lemmetized_data
```

```python
POS=[]
for data in tokenized_data:
  POS.append(nltk.pos_tag(data))
POS
```

```python
NER_data = []
for tagged_tokens in POS:
    ner_tree = nltk.ne_chunk(tagged_tokens)
    NER_data.append(ner_tree)

for i, tree in enumerate(NER_data):
    print(f"Sentence {i+1}:")
    print(tree)
    print("-" * 20)
```

```python
import pandas as pd
all_tokens = []
all_lemmatized = []
all_stemmed = []
all_pos = []

for i in range(len(tokenized_data)):
    original_tokens = tokenized_data[i]
    lemmatized_tokens = lemmetized_data[i]
    stemmed_tokens = stemmed_data[i]
    pos_tags = POS[i]
    min_sentence_len = min(len(original_tokens), len(lemmatized_tokens), len(stemmed_tokens), len(pos_tags))
    for j in range(min_sentence_len):
        all_tokens.append(original_tokens[j])
        wordnet_lemmatizer = nltk.stem.WordNetLemmatizer()
        porter_stemmer = nltk.stem.PorterStemmer()
        lemmatized_on_original = [wordnet_lemmatizer.lemmatize(word) for word in original_tokens]
        stemmed_on_original = [porter_stemmer.stem(word) for word in original_tokens]
        pos_on_original = nltk.pos_tag(original_tokens)


        all_lemmatized.append(lemmatized_on_original[j])
        all_stemmed.append(stemmed_on_original[j])
        all_pos.append(pos_on_original[j][1])
is_stopword = [token.lower() in stop_words for token in all_tokens]

data_for_csv = {
    'Original Token': all_tokens,
    'Is Stopword': is_stopword,
    'Lemmatized Token': all_lemmatized,
    'Stemmed Token': all_stemmed,
    'POS Tag': all_pos
}

df_tokens = pd.DataFrame(data_for_csv)

# Save the DataFrame to a CSV file
df_tokens.to_csv('processed_text_data_tokens.csv', index=False)

print("Processed data saved to processed_text_data_tokens.csv")
display(df_tokens.head())
```
